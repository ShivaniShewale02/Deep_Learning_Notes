# Generative Adversarial Network (GAN)

### What is a GAN?
- A GAN (Generative Adarial Network) is a deep learning model where two neural networks compete with each other — one generates fake data and the other tries to detect it — and through this competition, the generator learns to create realistic data.
- A Generative Adversarial Network is a generative deep learning model consisting of two neural networks — a Generator and a Discriminator — trained in an adversarial manner.
The Generator tries to generate realistic data from random noise, while the Discriminator tries to distinguish real data from fake data.
- Through this minimax game, the Generator learns the real data distribution and produces highly realistic outputs.
- GAN is a deep learning model where a generator creates fake data and a discriminator evaluates it, and both compete in a minimax game until the generated data becomes realistic.
