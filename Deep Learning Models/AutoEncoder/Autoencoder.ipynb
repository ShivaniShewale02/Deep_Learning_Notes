# Autoencoders

### What is an Autoencoder?

- An Autoencoder is a neural network trained to copy its input to its output.
- 🎯 Goal: Learn a compressed (latent) representation of data without labels
- It is used for task like unsupervised learning and dimensionality reduction tasks, its primarily desiged to learn efficient representation of data by encoding it into a lower dimensional latent space and then decoding it back to the original data spance.
- Autoencodered are widely used for various applications, including data compression, denoising, feature learning and anomaly detection.

It works by:
- Compressing input → Encoder
- Reconstructing input → Decoder

# Implementaion

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
```

```python
X = np.random.rand(1000, 20)  # 1000 samples, 20 features
```

```python
input_dim = X.shape[1]

input_layer = Input(shape=(input_dim,))
encoded = Dense(10, activation='relu')(input_layer)
encoded = Dense(5, activation='relu')(encoded)  # bottleneck
```

```python
decoded = Dense(10, activation='relu')(encoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)
```

```python
autoencoder = Model(input_layer, decoded)  # it is to decode the input 
```

```python
autoencoder.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
autoencoder.fit(
    X, X,                     # input = output
    epochs=10,
    batch_size=32,
    validation_split=0.2
)
```
