# Recuring Neural Network (RNN)

- A Recurrent Neural Network (RNN) is a class of neural networks designed to handle sequential data by maintaining a memory (hidden state) that captures information from previous time steps.
- Unlike Feedforward Neural Networks, RNNs reuse the same weights at each time step, allowing them to model temporal dependencies.
- RNNs solve sequence-based problem by remembering past information.



### RNN Architecture





| Symbol     | Meaning         |
| ---------- | --------------- |
| (x_t)      | Input at time t |
| (h_t)      | Hidden state    |
| (y_t)      | Output          |
| (W_{xh})   | Input → Hidden  |
| (W_{hh})   | Hidden → Hidden |
| (W_{hy})   | Hidden → Output |
| (b_h, b_y) | Biases          |

# Implementation

```python
import torch
import torch.nn as nn

class SimpleRNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        self.Wxh = nn.Linear(input_size, hidden_size)
        self.Whh = nn.Linear(hidden_size, hidden_size)
        self.Why = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h = torch.zeros(x.size(0), self.Whh.out_features)
        outputs = []

        for t in range(x.size(1)):
            h = torch.tanh(self.Wxh(x[:, t]) + self.Whh(h))
            y = self.Why(h)
            outputs.append(y)

        return torch.stack(outputs, dim=1)
```

# Sequence Creation for RNN

- Sequence creation means converting raw data into ordered chunks (sequences) so that an RNN can learn patterns over time.
- Sequence creation is the process of converting ordered data into fixed-length sequences so that RNNs can learn temporal dependencies.

🔑 RNNs don’t learn from single data points.
- They learn from ordered sequences of data.
