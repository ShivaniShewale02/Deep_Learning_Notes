### Model Compiling While Training 
- Compiling a model configures the loss function, optimizer, and evaluation metrics, defining how the network learns and how its performance is measured.

##### What does compiling a model mean?

👉 Compiling a model = configuring how the model will learn. 
When you build a neural network, you only define:

- Layers
- Connections
- Forward pass (structure)

But the model still does NOT know:

- ❓ How wrong it is
- ❓ How to update weights
- ❓ What to optimize

##### model.compile() answers these questions.

##### Why do we compile a model?

We compile a model to tell it three critical things:

| What              | Why                         |
| ----------------- | --------------------------- |
| **Loss function** | How bad the prediction is   |
| **Optimizer**     | How to update weights       |
| **Metrics**       | How to evaluate performance |

Without compilation → training cannot start.

# Comparison Table (Quick Selection Guide)

| Architecture | Key Strength         | Weakness         | Best Use            |
| ------------ | -------------------- | ---------------- | ------------------- |
| VGG          | Simple, deep         | Huge params      | Learning            |
| ResNet       | Stable deep training | Heavy            | General vision      |
| Inception    | Multi-scale features | Complex          | Scene understanding |
| DenseNet     | Feature reuse        | Memory heavy     | Medical imaging     |
| MobileNet    | Lightweight          | Lower accuracy   | Mobile/Edge         |
| EfficientNet | Best efficiency      | Slightly complex | Production          |
| U-Net        | Pixel-level output   | Task-specific    | Segmentation        |
| YOLO         | Real-time detection  | Lower precision  | Live detection      |


#### 🔹 1. Task Type
| Task             | Architecture          |
| ---------------- | --------------------- |
| Classification   | ResNet / EfficientNet |
| Segmentation     | U-Net                 |
| Detection        | YOLO / Faster R-CNN   |
| Mobile inference | MobileNet             |


#### 🔹 2. Dataset Size
| Dataset      | Recommendation               |
| ------------ | ---------------------------- |
| Small (<10k) | DenseNet / Transfer learning |
| Medium       | ResNet                       |
| Large        | EfficientNet                 |


#### 🔹 3. Hardware Constraints
| Hardware      | Choose                |
| ------------- | --------------------- |
| GPU / Cloud   | ResNet / EfficientNet |
| Mobile / Edge | MobileNet             |
| Low memory    | EfficientNet          |


#### 🔹 4. Accuracy vs Speed
| Priority     | Choose           |
| ------------ | ---------------- |
| Max accuracy | EfficientNet     |
| Balance      | ResNet           |
| Speed        | MobileNet / YOLO |
