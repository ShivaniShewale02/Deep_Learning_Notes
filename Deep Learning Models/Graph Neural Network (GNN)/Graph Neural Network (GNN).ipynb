# Graph Neural Network (GNN)

 - A Graph Neural Network (GNN) is a neural network designed to work on graph-structured data.
 - Instead of grids (images) or sequences (text), GNNs learn from nodes and their relationships (edges).

### Why Do We Need GNNs?

- Traditional neural networks fail on graph data because:

| Data Type              | Model             |
| ---------------------- | ----------------- |
| Images (grid)          | CNN               |
| Text / Time series     | RNN / Transformer |
| **Graphs (irregular)** | ❌ CNN/RNN fail    |

 Graphs are:
- Non-Euclidean
- Variable size
- No fixed ordering

👉 GNNs solve this.

### Simple Intuition

- A node learns by looking at its neighbors repeatedly.

Example:
- You understand a person better by knowing their friends
- After many rounds, you understand the whole network

### Graph Basics

##### Graph = (V, E)

| Symbol | Meaning                                 |
| ------ | --------------------------------------- |
| **V**  | Nodes (users, molecules, products)      |
| **E**  | Edges (friendship, bonds, interactions) |
| **X**  | Node features                           |
| **A**  | Adjacency matrix                        |


Graph → GNN Layer(s) → Node Embeddings → Task Head

### GNN Working Principle
Every GNN layer has 3 steps:

##### 1️⃣ Message Passing
- Each node collects information from neighbors

##### 2️⃣ Aggregation
- Neighbor messages are combined

##### 3️⃣ Update
- Node updates its embedding
