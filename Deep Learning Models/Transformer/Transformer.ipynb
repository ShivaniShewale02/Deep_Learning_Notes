# Transformers

- Transformers are the backbone of modern Deep Learning, especially in NLP, Vision, Multimodal AI, and Generative AI. They replaced RNNs/LSTMs by enabling parallel processing, long-range dependency learning, and scalability.
- Transformers are a deep learning architecture introduced in 2017 that use self-attention mechanisms to process sequential data like text, images, and audio — without using RNNs or CNNs.
- They power modern AI systems like ChatGPT, BERT, Google Translate, and Vision Transformers.
- Transformers are attention-based neural networks that understand relationships between all elements in a sequence at once.
- Transformers use self-attention to model global dependencies efficiently without recurrence, enabling scalable deep learning models across multiple domains.

### Interview Answers

- Transformers are deep learning models based on self-attention mechanisms that process sequences in parallel and capture long-range dependencies efficiently. They replaced RNNs in NLP tasks and power modern LLMs like GPT and BERT.
- Transformers are neural network architectures introduced in the paper “Attention is All You Need.” They rely on self-attention instead of recurrence or convolution to understand relationships between words in a sequence. They consist of encoder and decoder blocks and allow parallel processing, better long-range dependency handling, and scalability, which makes them ideal for large language models like GPT and BERT.

### Simple Explanation (Non-Technical)

- Imagine you are reading a sentence:

“The dog chased the cat because it was scared.”

- To understand what “it” refers to, your brain looks at all words in the sentence at once.
- That’s exactly what a Transformer does.

👉 Instead of reading word by word like an RNN, it:
- Looks at all words simultaneously
- Learns which words are important for each other
- Builds a deep understanding of context

### Why Transformers Were Introdued?

Problems with RNN / LSTM

| Issue                     | Explanation               |
| ------------------------- | ------------------------- |
| Sequential processing     | Cannot parallelize → slow |
| Long-term dependency loss | Vanishing gradients       |
| Poor scalability          | Hard for large datasets   |

Transformers solve this using Attention.

##### Core Idea: Attention Is All You Need
- The original Transformer paper (2017) showed that attention alone is sufficient, without recurrence or convolution.

## Transformer Architecture (High-Level)

##### Two Main Blocks
1. Encoder → Understands input
2. Decoder → Generates output

Used together in:
- Translation
- Summarization
- Question Answering

Used alone in:
- BERT (Encoder-only)
- GPT (Decoder-only)

Some models use:
- Encoder only (like BERT)
- Decoder only (like GPT)
- Both (like T5)

## Transformer Encoder (Deep Dive)

- A Transformer Encoder converts an input sequence (text, image patches, tokens) into a rich contextual representation.
- 👉 It answers: “What does this input mean?”

🔹 Encoder Stack (Repeated N Times)

Each encoder layer contains:
- Multi-Head Self-Attention
- Add & Layer Normalization
- Feed Forward Neural Network
- Add & Layer Normalization


🔸 Encoder Workflow (Step-by-Step)
1. Intut Embedding
2. Add Positional Encoding
3. Self-Attention (Q=K=V)
4. Feed Forword Network
5. Output = Contextual Embeddings

## Transformer Decoder (Deep Dive)

- A Transformer Decoder generates an output sequence, one token at a time.
- 👉 It answers: “What should I generate next?”

##### Decoder Layers Contain:
- Masked Self-Attention
- Encoder-Decoder Attention
- Feed Forward Network

#### 🔸 Decoder Workflow
1. Take previous output tokens
2. Apply masked self-attnetion
3. Attend to encoder output
4. Generate next token


#### 🔹 Masked Attention
- Prevents future tokens from being seen (🚫 Decoder cannot see future tokens.)

- Enables auto-regressive generation

Used in:
- Text generation
- Language modeling
