# Attention

Attention is a mechanism that allows a model to:
- Focus on the most relevant parts of input while processing a specific element.
- Instead of treating all words equally, it assigns importance weights.

### Human Analogy

When you read:
- “The animal didn’t cross the road because it was tired.”
- To understand “it”, you focus more on animal than “road”.

👉 That focusing mechanism = Attention

### 🔥 Types of Attention
1️⃣. Global Attention
- Looks at all input tokens

2️⃣. Local Attention
- Looks at nearby tokens

3️⃣. Additive Attention (Bahdanau)
- Uses feedforward network

4️⃣. Multiplicative Attention (Luong)
- Uses dot-product similarity

- Basically we check the Query (Q) with each Key (K) to understand how similar is the key to the query, now when we find the matching key then the job is to extract the Value (V) for that key or for the Keys we found most similar.
- We compute Attention Score; compute pairwise similarity between each query and key: with dot product of Q with K / scale it (scaling) : cosine similarity basically.

# Self-Attention
### What is Self Attention?
- Now comes the code of Transformer.
- Self-Attention is when, a sequence attends to itself to compute contextual representation.
- Meaning, Each word looks at all other words in same sentance.
