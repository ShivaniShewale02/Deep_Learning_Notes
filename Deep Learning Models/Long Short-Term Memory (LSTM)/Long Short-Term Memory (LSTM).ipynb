# What is LSTM?

- LSTM (Long Short-Term Memory) is a special type of RNN that can remember important information for a long time and forget useless information.
- LSTM (Long Short-Term Memory) is a special type of Recurrent Neural Network (RNN) designed to learn long-term dependencies in sequential data.
- LSTM solves Vanishing Gradient and poor long term memmory problem using a memory cell and gating mechanism.
- LSTM uses a memory cell and gates to decide what to remember, what to forget, and what to output.
- LSTM is a type of RNN designed to learn long-term dependencies. It uses a memory cell and three gates—forget, input, and output—to control information flow. This structure prevents vanishing gradients and allows the model to remember important information over long sequences.
- LSTM is an RNN with controlled memory that decides what to remember, forget, and output over time.

### 🧠 Real-Life Intuition (Best Way to Remember)

Imagine you are reading a story 📖

You don’t remember every word, but you remember:
- Important names
- Important events
- Forget useless details

👉 LSTM works exactly like this.

## Architecture



#### Components:
| Symbol   | Meaning               |
| -------- | --------------------- |
| **xₜ**   | Input at time t       |
| **hₜ₋₁** | Previous hidden state |
| **Cₜ₋₁** | Previous cell state   |
| **hₜ**   | Current hidden state  |
| **Cₜ**   | Updated cell state    |

Each LSTM cell has:
- Cell State (C) → long-term memory
- Hidden State (h) → short-term output
- Three Gates → controllers

## What is Cell State?
🧠 Cell State = Memory notebook
- Flows almost straight
- Information changes slowly
- Prevents forgetting

📌 This is why LSTM avoids vanishing gradients.



## LSTM Gates
- Think of gates as decision makers (YES / NO).

### 1️⃣ Forget Gate – What should I forget?
- Decides what old information should be removed
Looks at:
- Previous memory
- Current input
- Decides what information is useless

Output: (Output between 0 and 1)
- 0 → forget
- 1 → keep

🧠 Example:
Forget yesterday’s irrelevant noise




### 2️⃣ Input Gate – What to learn?
- Decides what new information should be stored
- Adds useful information to memory

Two steps:
- a) Candidate Memory (New possible information)

- b) Input Gate (Decides how much to store)

🧠 Example:
Store country name “France”




### 3️⃣ Cell State Update (Long-term memory - Memory update)
- Old memory + new important info

Two steps:
- a) Candidate Memory (New possible information)
- b) Input Gate (Decides how much to store)

🧠 Like:
Erase some lines, write new lines




### 4️⃣ Output Gate (What to output? - What should I say now?)
- Decides what part of memory to expose (use as output)
- Produces hidden state (output)

Two steps:
- a) Candidate Memory (New possible information)
- b) Input Gate (Decides how much to store)

🧠 Example:
Say the correct next word
