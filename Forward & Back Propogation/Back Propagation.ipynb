# Back Propagation

### What is Backpropagation?
Backpropagation (short for backward propagation of errors) is the learning algorithm used in Artificial Neural Networks (ANNs) to update weights and biases based on the error from predictions.

It uses gradient descent optimization and the chain rule of calculus to minimize the cost function.

Backpropagation (short for “backward propagation of errors”) is the algorithm used to train neural networks by adjusting weights and biases to minimize the loss function.

It works in two phases:
- Forward Pass → Input flows through the network to compute the output (prediction).
- Backward Pass → Error (difference between predicted and actual output) flows backward, updating weights using gradient descent.

### Why Backpropagation?
- Neural networks have multiple layers (hidden + output).
- Errors occur at the output, but to adjust weights in earlier layers, we need to propagate the error backward.
- Backpropagation tells us how much each weight contributed to the error.
